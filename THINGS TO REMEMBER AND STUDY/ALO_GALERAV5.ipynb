```python

base = sqlite3.connect("/content/drive/MyDrive/DataBase/AventuraPoste.db")
database = base.cursor()
df = pd.read_sql("SELECT * FROM Poste", base, parse_dates=['Data'], index_col='Data')
#df = df[df.index.year == 2019]
df['year'] = df.index.year
df['month'] = df.index.month
df = df[(df.index.year == 2021) | (df.index.year == 2020)]
df['year'] = df.index.year
df['month'] = df.index.month
df = (df.groupby(['year','Nome', 'month']))['month'].count().unstack()
df.columns = ['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho', 'Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro']


df = df[df.index.get_level_values(0) == 2021].loc[2021]
top10 = df.sum(axis=1).sort_values(ascending=False)[:11]
top10 = top10.drop("Redação")


fig, ax = plt.subplots(figsize=(20, 12))
for nome in top10.index:
  df[df.index == nome].iloc[0].plot(marker='o', legend=True, ax=ax)

#ESTILO
plt.style.use("dark_background")

for param in ['text.color', 'axes.labelcolor', 'xtick.color', 'ytick.color']:
  plt.rcParams[param] = '0.9'  # very light grey

for param in ['figure.facecolor', 'axes.facecolor', 'savefig.facecolor']:
  plt.rcParams[param] = '#212946'  # bluish dark grey

ax.grid(color='#2A3459')

ax.set_xlim([ax.get_xlim()[0] - 0.2, ax.get_xlim()[1] + 0.2])  # to not have the markers cut off
ax.set_ylim(0)
ax.set_xlabel("Mêses")
ax.set_ylabel("Postes")
ax.set_xticks([x for x in range(0, 12)])
ax.set_xticklabels(df.columns)
ax.legend(title='Autores')

ax.set_title("FREQUENCIA DE PUBLICAÇOES POR MÊS DOS PRINCIPAIS AUTORES 2021")
#ESTILO



```

```python
base = sqlite3.connect("/content/drive/MyDrive/DataBase/AventuraPoste.db")
database = base.cursor()
df = pd.read_sql("SELECT * FROM Poste", base, parse_dates=['Data'], index_col='Data')
#df = df[df.index.year == 2019]
df['year'] = df.index.year
df['month'] = df.index.month
df = df[(df.index.year == 2021)]
df['year'] = df.index.year
df['month'] = df.index.month


dfmateria = df[df['Categoria'] == 'Materias']
dfnoticia = df[df['Categoria'] == 'Noticias']
dfcuriosidade = df[df['Categoria'] == 'Curiosidade']
dfcorona = df[df['Categoria'] == 'CoronaVirus']

#df['Nome'][df['Categoria'] == "Materias"].value_counts()
dfmateria = (dfmateria.groupby(['Nome','month']))['Categoria'].count().unstack()
dfmateria.columns = ['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho', 'Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro']

dfnoticia = (dfnoticia.groupby(['Nome','month']))['Categoria'].count().unstack()
dfnoticia.columns = ['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho', 'Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro']

dfcuriosidade = (dfcuriosidade.groupby(['Nome','month']))['Categoria'].count().unstack()
dfcuriosidade.columns = ['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho', 'Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro']

dfcorona = (dfcorona.groupby(['Nome','month']))['Categoria'].count().unstack()
dfcorona.columns = ['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho', 'Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro']

dfmateria = dfmateria.fillna(0)
dfnoticia = dfnoticia.fillna(0)
dfcuriosidade = dfcuriosidade.fillna(0)
dfcorona = dfcorona.fillna(0)


df = dfmateria*2
df = df.add(dfnoticia, fill_value=0)
df = df.add(dfcuriosidade*1.5, fill_value=0)
df = df.add(dfcorona, fill_value=0)

df = df.drop("Redação")


top10 = df.sum(axis=1).sort_values(ascending=False)[:10]

#print(top10)

#print(dfmateria[dfmateria.index == 'Isabela Barreiros'])


fig, ax = plt.subplots(dpi=100,figsize=(25, 12))
for nome in top10.index:
  df[df.index == nome].iloc[0].plot(marker='o', legend=True, ax=ax)

fig2, ax2 = plt.subplots(dpi=100,figsize=(15, 5))
top10.plot(kind="barh", ax=ax2)
for index, y in enumerate(top10):
  ax2.text(y, index + 0.15 , str(y), fontweight='bold')

ax2.grid(b = True, color ='grey',
        linestyle ='-.', linewidth = 0.5,
        alpha = 0.2)

ax2.invert_yaxis()
ax2.set_title("PONTUAÇÂO DOS PRINCIPAIS AUTORES 2021")
ax2.set_xlabel("Pontuação")

#ESTILO
plt.style.use("dark_background")

for param in ['text.color', 'axes.labelcolor', 'xtick.color', 'ytick.color']:
  plt.rcParams[param] = '0.9'  # very light grey

for param in ['figure.facecolor', 'axes.facecolor', 'savefig.facecolor']:
  plt.rcParams[param] = '#212946'  # bluish dark grey

ax.grid(color='#2A3459')
ax.set_xlim([ax.get_xlim()[0] - 0.2, ax.get_xlim()[1] + 0.2])  # to not have the markers cut off
ax.set_ylim(0)
ax.set_xlabel("Mêses")
ax.set_ylabel("Pontuação")
ax.set_xticks([x for x in range(0, 13)])
ax.set_xticklabels(df.columns)
ax.legend(title='Autores')
ax.set_title("PONTUAÇÂO DOS PRINCIPAIS AUTORES POR MÊS 2021")
#ESTILO



```
